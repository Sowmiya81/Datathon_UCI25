```python
import pandas as pd
```

```python
df = pd.read_csv('cleaned.csv', index_col=0)
df.head()
```

```python
df_new = df.dropna().copy()
df_new
```

```python
from gensim.utils import simple_preprocess

# Tokenize relevant columns
df_new['prompt_tokens'] = df_new['prompt_valuable'].apply(lambda x: simple_preprocess(str(x)))
df_new['response_a_tokens'] = df_new['response_a_cleaned'].apply(lambda x: simple_preprocess(str(x)))
df_new['response_b_tokens'] = df_new['response_b_cleaned'].apply(lambda x: simple_preprocess(str(x)))
```

```python
# Print original and tokenized text for the first 3 rows
for i in range(3):
    print(f"\nRow {i+1}")
    print("="*40)
    print("Original Prompt:")
    print(df_new['prompt'].iloc[i])
    print("Tokenized Prompt:")
    print(df_new['prompt_tokens'].iloc[i])
    
    print("\nOriginal Response A:")
    print(df_new['response_a'].iloc[i])
    print("Tokenized Response A:")
    print(df_new['response_a_tokens'].iloc[i])
    
    print("\nOriginal Response B:")
    print(df_new['response_b'].iloc[i])
    print("Tokenized Response B:")
    print(df_new['response_b_tokens'].iloc[i])
    print("="*40)

```

```python
# Combine all tokenized texts to form a training corpus
tokenized_corpus = df_new['prompt_tokens'].tolist() + df_new['response_a_tokens'].tolist() + df_new['response_b_tokens'].tolist()
# Print the number of documents in the corpus
print(f"Total documents in tokenized corpus: {len(tokenized_corpus)}")

# Print a few sample tokenized documents
print("\nSample tokenized documents from corpus:")
for i, doc in enumerate(tokenized_corpus[:5]):  # Change 5 to any number of samples you want to see
    print(f"\nDocument {i+1}:")
    print(doc)
```

```python
# Step 1: Import necessary libraries
from gensim.models import Doc2Vec
from gensim.models.doc2vec import TaggedDocument
from sklearn.metrics.pairwise import cosine_similarity

# Step 2: Prepare Tagged Documents
tagged_docs = [
    TaggedDocument(words=row['prompt_tokens'], tags=[f"PROMPT_{i}"]) 
    for i, row in df_new.iterrows()
] + [
    TaggedDocument(words=row['response_a_tokens'], tags=[f"RESPONSE_A_{i}"]) 
    for i, row in df_new.iterrows()
] + [
    TaggedDocument(words=row['response_b_tokens'], tags=[f"RESPONSE_B_{i}"]) 
    for i, row in df_new.iterrows()
]

# Step 3: Train the Doc2Vec model
doc2vec_model = Doc2Vec(
    vector_size=100,  # Size of the vector
    window=5,         # Context window
    min_count=2,      # Ignore words with fewer than 2 occurrences
    workers=4,        # Number of threads to run in parallel
    epochs=30         # Number of training epochs
)

# Build the vocabulary and train the model
doc2vec_model.build_vocab(tagged_docs)
doc2vec_model.train(tagged_docs, total_examples=doc2vec_model.corpus_count, epochs=doc2vec_model.epochs)

# Step 4: Get Document Vectors for each response (in this case, 'response_a' and 'response_b')
df_new['response_a_vector'] = df_new['response_a_tokens'].apply(lambda x: doc2vec_model.infer_vector(x))
df_new['response_b_vector'] = df_new['response_b_tokens'].apply(lambda x: doc2vec_model.infer_vector(x))
```

```python
# Step 5: Displaying Document Vectors for 'response_a' and 'response_b'
print("\nDisplaying response vectors for the first few rows:")
for i in range(5):  # Change 5 to display more rows
    print(f"Response A Vector for row {i}:")
    print(df_new['response_a_vector'][i])
    print(f"Response B Vector for row {i}:")
    print(df_new['response_b_vector'][i])
    print()

# Step 6: Displaying similarity scores for each response pair
df_new['similarity'] = df_new.apply(
    lambda row: cosine_similarity([row['response_a_vector']], [row['response_b_vector']])[0][0], axis=1
)

# Print similarity scores along with corresponding responses
print("\nSimilarity scores for response pairs:")
print(df_new[['response_a', 'response_b', 'similarity']].head())

# Optional: Perform a similarity search for a specific document (e.g., prompt)
query_doc = df_new['prompt_tokens'][0]  # Change to any document from 'prompt_tokens'
query_vector = doc2vec_model.infer_vector(query_doc)

# Find the most similar documents based on the query
similar_docs = []
for i, row in df_new.iterrows():
    doc_vector = doc2vec_model.infer_vector(row['prompt_tokens'])
    similarity_score = cosine_similarity([query_vector], [doc_vector])[0][0]
    similar_docs.append((i, similarity_score))

# Sort by similarity score and display top 5 most similar documents
similar_docs.sort(key=lambda x: x[1], reverse=True)
print("\nTop 5 most similar documents to the query prompt:")
for doc_id, score in similar_docs[:5]:
    print(f"\nDocument {doc_id}: Similarity Score: {score}")
    print(f"Prompt: {df_new['prompt'][doc_id]}")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Step 1: Calculate the cosine similarity between 'response_a' and 'response_b' for each row
df_new['similarity'] = df_new.apply(
    lambda row: cosine_similarity([row['response_a_vector']], [row['response_b_vector']])[0][0], axis=1
)

# Step 2: Save the updated dataframe to a new CSV file
df_new.to_csv('updated_similarities.csv', index=False)

# Optionally, print the first few rows of the dataframe to verify the new column
print(df_new[['response_a', 'response_b', 'similarity']].head())
```
